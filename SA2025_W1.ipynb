### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())

median_mpg = df['mpg'].median()

cars_morethan_median = df[df['mpg'] >= median_mpg]

cars_morethan_median['hp_to_wt'] = cars_morethan_median['horsepower'] / cars_morethan_median['weight']

cars_with_best_hp_to_wt = cars_morethan_median.sort_values(by= 'hp_to_wt', ascending= False)

cars_with_best_hp_to_wt.head()
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
df_group = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

for origin in df_group['origin'].unique():
    data = df_group[df_group['origin'] == origin]
    plt.plot(data['model_year'], data['mpg'], label=origin)

plt.title('Average MPG by Model Year and Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')

# Add a legend to distinguish the origins
plt.legend(title='Origin')

plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
df_1 = pd.read_csv('Cars.csv')

name_counts = df_1['name'].value_counts()

multiple_appearance_names = name_counts[name_counts > 1].index

consistent_cars = df_1[df_1['name'].isin(multiple_appearance_names)]

mpg_std = consistent_cars.groupby('name')['mpg'].std()

low_variation_cars = mpg_std[mpg_std < 1.0].index

final_consistent_cars = consistent_cars[consistent_cars['name'].isin(low_variation_cars)]

final_cars = final_consistent_cars.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    average_mpg=('mpg', 'mean')
).reset_index()

final_result = final_cars.sort_values(by=['appearances', 'average_mpg'], ascending=[False, False])

print(final_result)
```
